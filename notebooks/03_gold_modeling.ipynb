```python
#Lendo documento da silver
import pandas as pd

df_silver = pd.read_parquet(
    r"D:\Geral\Projetos\hospital-readmission-data-engineering\data\silver\hospital_readmissions_clean.parquet"
)
```

```python
#Criando tabelas da gold
dim_patient = df_silver[["patient_id", "age", "age_group", "gender"]].drop_duplicates()
dim_patient.to_parquet(r"..\data\gold\dim_patient.parquet", index=False)

dim_clinical = df_silver[["patient_id", "bmi", "cholesterol", "systolic_pressure", "diastolic_pressure", "blood_pressure_stage", "diabetes", "hypertension", "metabolic_risk"]].drop_duplicates()
dim_clinical.to_parquet(r"..\data\gold\dim_clinical.parquet", index=False)

dim_discharge = df_silver[["patient_id", "discharge_destination"]].drop_duplicates()
dim_discharge.to_parquet(r"..\data\gold\dim_discharge.parquet", index=False)

fact_readmission = df_silver[["admission_id", "patient_id", "length_of_stay", "medication_count", "readmitted_30_days"]]
fact_readmission.to_parquet(r"..\data\gold\fact_readmission.parquet", index=False)
```
