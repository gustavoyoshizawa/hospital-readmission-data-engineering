```python
import pandas as pd
import os

#Caminho do dataset
file_path = r"D:\Geral\Projetos\hospital-readmission-data-engineering\kagglehub\datasets\siddharth0935\hospital-readmission-predictionsynthetic-dataset\versions\1"
csv_path = os.path.join(file_path, "hospital_readmissions_30k.csv")

#Ler csv
df = pd.read_csv(csv_path, sep=",")


#Criando patsa bronze
os.makedirs(r"..\data\bronze", exist_ok=True)

#Salvando como parquet
df.to_parquet(r"..\data\bronze\hospital_readmissions.parquet", index=False)

print("✅ Bronze layer criada com sucesso!")
print(f"Linhas: {df.shape[0]} | Colunas: {df.shape[1]}")
```

```python
# Novo caminho para o Excel
excel_path = os.path.join(file_path, "hospital_readmissions_30k.xlsx")

# Exportar para Excel
df.to_excel(excel_path, index=False)
```
